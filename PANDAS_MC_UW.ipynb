##Ejercicio 1
Porcentajes: Juan tiene 5 manzanas, 3 naranjas y 1 pera. Pedro tiene 2
manzanas, 4 naranjas y 6 peras.Ana tiene 5 manzanas, 2 naranjas y 8
peras ¿Qué porcentaje de frutas tienen en total?.

```python
import pandas as pd
nombres = ['Juan', 'Pedro', 'Ana']
manzanas = [5, 2, 5]
naranjas = [3, 4, 2]
peras = [1, 6, 8]
df = pd.DataFrame({
    'NOMBRES': nombres,
    'MANZANAS': manzanas,
    'NARANJAS': naranjas,
    'PERAS': peras,
})
df

df['TOTAL_FRUTAS'] = df['MANZANAS'] + df['NARANJAS'] + df['PERAS']

df['PORCENTAJE'] = (df['TOTAL_FRUTAS'] / df['TOTAL_FRUTAS'].sum()) * 100

print("Porcentaje de frutas en total por persona:")
df[['NOMBRES', 'PORCENTAJE']]
```

##Ejercicio 2
Escribir una función que reciba un diccionario con las notas de los alumno de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import pandas as pd
import numpy as np
alumnos = ['Juan', 'Pedro', 'Ana']
notas = [ 13, 17, 10]
df = pd.DataFrame({
    'ALUMNOS': alumnos,
    'NOTAS': notas,
})
df

nmin = df['NOTAS'].min()
nmax = df['NOTAS'].max()
media = df['NOTAS'].mean()
desv_tip = df['NOTAS'].std()

stats = pd.Series({
    'Mínima': nmin,
    'Máxima': nmax,
    'Media': media,
    'Desviación Típica': desv_tip
})
stats
```

##Ejercicio 3
Escribir una función que reciba un diccionario con las notas de los
alumnos de un curso y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.

```python
import pandas as pd
import numpy as np
alumnos = ['Juan', 'Pedro', 'Ana']
notas = [ 13, 17, 10]
df = pd.DataFrame({
    'ALUMNOS': alumnos,
    'NOTAS': notas,
})
df

naprobadas = df[df['NOTAS']>=11]
naprobadas

aprobados = naprobadas['NOTAS'].sort_values(ascending=False)
aprobados
```

##Ejercicio 4
Escribir un programa que genere y muestre por pantalla un DataFrame con los datos de la tabla siguiente:

Mes Ventas Gastos

Enero 30500 22000

Febrero 35600 23400

Marzo 28300 18100

Abril 33900 20700

```python
import pandas as pd
mes = ['Enero', 'Febrero', 'Marzo', 'Abril']
ventas = [30500, 35600, 28300, 33900]
gastos = [22000, 23400, 18100, 20700]
df = pd.DataFrame({
    'MES': mes,
    'VENTAS': ventas,
    'GASTOS': gastos,
})
df
```

##Ejercicio 5
Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos)
total en los meses indicados.

```python
meses = df[df['MES'].isin(['Enero'])]

bal_ventas = meses['VENTAS'].sum()
bal_gastos = meses['GASTOS'].sum()

bal_total = bal_ventas - bal_gastos

print('Balance:', bal_total)
```
