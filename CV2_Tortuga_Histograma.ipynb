```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
imagen=cv2.imread("/content/tortuga.png")
canales = cv2.split(imagen)
colores = ('b', 'g', 'r')
plt.figure()
plt.title('Histograma')
plt.xlabel('Bits')
plt.ylabel('#Pixeles')
for (canal, color) in zip(canales,colores):
    hist = cv2.calcHist([canal], [0], None, [256], [0, 256])
    plt.plot(hist, color = color)
    plt.xlim([0, 256])
plt.show()
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/tortuga.png',1)
# Crear máscara
mask = np.zeros(img.shape[:2], np.uint8)
mask[50:150, 50:200] = 255
masked_img = cv2.bitwise_and(img,img,mask = mask)
# Calcular el histograma con máscara y sin máscara
# Fijar el tercer argumento como "mask"
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])
plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# mostrar el histograma en gris
img = cv2.imread('/content/tortuga.png')
# cambiar la imagen como vector y mostrar como barras
plt.hist(img.ravel(),256,[0,256]);
plt.show()
# Descomponiendo imagen a color para histograma
imagen=cv2.imread('/content/tortuga.png')
canales = cv2.split(imagen)
colores = ('b', 'g', 'r')
plt.figure()
plt.title('Histograma')
plt.xlabel('Bits')
plt.ylabel('#Pixeles')
for (canal, color) in zip(canales,colores):
    hist = cv2.calcHist([canal], [0], None, [256], [0, 256])
    plt.plot(hist, color = color)
    plt.xlim([0, 256])
plt.show()
# utilizando opencv
img = cv2.imread('/content/tortuga.png')
# Crear máscara
mask = np.zeros(img.shape[:2], np.uint8)
mask[30:130, 100:150] = 255
masked_img = cv2.bitwise_and(img,img,mask = mask)

# Calcular el histograma con máscara y sin máscara
# Fijar el tercer argumento como "mask"
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```
